```python
import sqlite3
import numpy as np

db_file_path = 'iris.db'

# データベースを読み込む
conn = sqlite3.connect(db_file_path)

# DBを操作するカーソルオブジェクトを作成
cur = conn.cursor()

# テーブル名を取得
cur.execute("SELECT name FROM sqlite_master WHERE type='table'")
print(cur.fetchall())
```

```python
# テーブルのカラム名を取得
cur.execute("SELECT name FROM PRAGMA_TABLE_INFO('iris')")
print(cur.fetchall())
```

```python
# データを表示
cur.execute('SELECT * FROM iris')

# 取得したデータを出力
for row in cur:
    print(row)
```

```python
# データ抽出。species が setosa で sepal_length が 3.0以下
for row in cur.execute("SELECT * FROM iris WHERE species == 'setosa' AND sepal_width <= 3.0"):
    print(row)
```

```python
# データ抽出。sepal_length が 6.5と7.0の間
for row in cur.execute("SELECT * FROM iris WHERE sepal_length BETWEEN 6.5 AND 7.0"):
    print(row)
```

```python
# データ抽出。petal_length が 3.0, 4.0, 5.0 のいずれかと等しい
for row in cur.execute("SELECT * FROM iris WHERE petal_length IN(3.0, 4.0, 5.0)"):
    print(row)
```

```python
# データ抽出。speciesname の中に sa を含む文字列
for row in cur.execute("SELECT * FROM iris WHERE species LIKE '%sa%'"):
    print(row)
```

```python
# petal_width の値を降順にソートする。昇順にする場合はDESCを除く
for row in cur.execute("SELECT * FROM iris ORDER BY petal_width DESC"):
    print(row)
```

```python
### 以降、csvファイルへ出力する手順
new_file_path = 'iris_extract_data.csv'

# テーブル名を取得
cur.execute("SELECT name FROM sqlite_master WHERE type='table'")
print(cur.fetchall())
```

```python
# テーブルのカラム名を取得
cur.execute("SELECT name FROM PRAGMA_TABLE_INFO('iris')")
print(cur.fetchall())
```

```python
# カラム名をリストで取得
cur.execute("SELECT name FROM PRAGMA_TABLE_INFO('iris')")
col_list = np.array(cur.fetchall()).flatten().tolist() # numpyの flatten を利用して展開して、それを再リスト化
col_list
```

```python
# csvファイルへ保存する
with open (new_file_path, 'w', encoding = 'utf-8') as f:
    col_str = ','.join(col_list)
    f.write(col_str + '\n')
    for row in cur.execute("SELECT * FROM iris WHERE species == 'setosa' AND sepal_width <= 3.0"):
        f.write(','.join([str(c) for c in row]) + '\n')
```

```python
# オブジェクトを閉じる
conn.close()
```
